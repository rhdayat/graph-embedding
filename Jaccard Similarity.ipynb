```python
data = data.drop(['Unnamed: 0','Ingredient','Manufacturers','Product',], axis=1)
data.head(4)
```

```python
def jaccard_similarity(x, y):
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
```

```python
jaccard_similarity(data['from'][0], data['to'][0])
```

```python
data['Jaccard'] = data.apply(lambda row: jaccard_similarity(row['from'], row['to']), axis=1)
```
